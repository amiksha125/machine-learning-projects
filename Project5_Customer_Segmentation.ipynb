# **Project 5: Customer Segmentation using K-Means Clustering**

Customer Segmentation means grouping customers based on their spending patterns and purchase behaviour.

As a data scientist in mall, you can build a system that can cluster customers into different groups, one group of customer may represent customers tends to purchase more in a month, and other group may show customers who do not purchase often in a month.

**Work Flow:** Customer Data --> Data pre processing  --> Data Analysis --> Optimal number of clusters(WCSS) --> K-Means Clustering Aldorithm --> Visualizing the Clusters

Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

Data Collection and Analysis

```python
# loading datafrom csv file to a Pandas Dataframe
customer_data = pd.read_csv('/content/Mall_Customers.csv')
```

```python
# first 5 rows in the data set
customer_data.head()
```

```python
# no of rows and columns in data set
customer_data.shape
```

```python
# getting some informations about the data set
customer_data.info()
```

```python
# checking for missing values
customer_data.isnull().sum() # if there  exist missingvalue, we can perform imputations
```

Choosing the Annual Income column & Spending Score

```python
# customer id, gender and age will not  be considered
X = customer_data.iloc[:, [3, 4]].values # index starts from 0 for columns
```

```python
print(X) # annual income and spending scores
```

Choosing the number of clusters

WCSS - Within Cluster Sum of Square

```python
# finding wcss value for different numbers of clusters

wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init= 'k-means++', random_state= 42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
# plot an elbow graph
sns.set()
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum number of clusters = 5

Training the K-Means Clustering Model

```python
kmeans = KMeans(n_clusters= 5, init='k-means++', random_state= 0)

# return a label for each data point based on their clusters
# each row will get its cluster number

Y = kmeans.fit_predict(X)
print(Y)
```

Visualizing all the clusters

5 Clusters - 0, 1, 2, 3, 4
X[Y == 0, 0] -> Y == 0 , taking all the values of cluster 1, and 0 means first column of X data

X[Y == 0, 1] -> Y == 0 , taking all the values of cluster 1, and 1 means second column of X data


```python
# plotting all the clusters with their centroids
plt.figure(figsize= (8, 8))
plt.scatter(X[Y == 0, 0], X[Y == 0, 1], s = 50, c = 'green', label = 'Cluster 1')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(X[Y == 2, 0], X[Y == 2, 1], s = 50, c = 'red', label = 'Cluster 3')
plt.scatter(X[Y == 3, 0], X[Y == 3, 1], s = 50, c = 'purple', label = 'Cluster 4')
plt.scatter(X[Y == 4, 0], X[Y == 4, 1], s = 50, c = 'yellow', label = 'Cluster 5')

# plot the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'cyan', label = 'Centroids')

plt.title('Customer Segmentations')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

We can clearly find the group of customers having annual income more but spending score less, we can target this group of customers to increase the sales and revenue
